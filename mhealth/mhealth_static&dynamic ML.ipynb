```python
activity = {
    0: 'Null',
    1: 'Standing still',
    2: 'Sitting and relaxing',
    3: 'Lying down',
    4: 'Walking',
    5: 'Climbing stairs',
    6: 'Waist bends forward',
    7: 'Frontal elevation of arms',
    8: 'Knees bending (crouching)',
    9: 'Cycling',
    10: 'Jogging',
    11: 'Running',
    12: 'Jump front & back'
}
```

```python
features = features.replace({'subject':'subject1'},1)
features = features.replace({'subject':'subject2'},2)
features = features.replace({'subject':'subject3'},3)
features = features.replace({'subject':'subject4'},4)
features = features.replace({'subject':'subject5'},5)
features = features.replace({'subject':'subject6'},6)
features = features.replace({'subject':'subject7'},7)
features = features.replace({'subject':'subject8'},8)
features = features.replace({'subject':'subject9'},9)
features = features.replace({'subject':'subject10'},10)
```

```python
features['target'] = features['activity'].copy()
```

```python
features.loc[features['target'].isin(['Standing still', 'Sitting and relaxing', 'Lying down']), 'target'] = 0
features.loc[features['target'].isin(['Walking', 'Climbing stairs', 'Waist bends forward', 'Frontal elevation of arms', 'Knees bending (crouching)',
                                      'Cycling', 'Jogging', 'Running', 'Jump front & back']), 'target'] = 1
```

```python
##x,y,z dont have the same variance so we need to scale the data 
#Standardized data:

from sklearn.preprocessing import StandardScaler

X = features[['acc_ch_y', 'acc_ch_z', 'acc_la_x', 'acc_la_y', 'acc_la_z', 'gyr_la_x',
       'gyr_la_y', 'gyr_la_z', 'mag_la_x', 'mag_la_y', 'mag_la_z', 'acc_rw_x',
       'acc_rw_y', 'acc_rw_z', 'gyr_rw_x', 'gyr_rw_y', 'gyr_rw_z', 'mag_rw_x',
       'mag_rw_y', 'mag_rw_z', 'subject']]
              
y= features['target']
              
scaler = StandardScaler()
X = scaler.fit_transform(X)

scaled_X = pd.DataFrame(data = X, columns = ['acc_ch_y', 'acc_ch_z', 'acc_la_x', 'acc_la_y', 'acc_la_z', 'gyr_la_x',
       'gyr_la_y', 'gyr_la_z', 'mag_la_x', 'mag_la_y', 'mag_la_z', 'acc_rw_x',
       'acc_rw_y', 'acc_rw_z', 'gyr_rw_x', 'gyr_rw_y', 'gyr_rw_z', 'mag_rw_x',
       'mag_rw_y', 'mag_rw_z', 'subject'])
scaled_X['label'] = y.values

scaled_X
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
scaled_X["label"] = encoder.fit_transform(scaled_X["label"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_X, 
                                                    scaled_X["label"],
                                                    test_size = 0.2, 
                                                    shuffle = True, 
                                                    random_state = 21)
```

```python
print("X_train.shape: ", X_train.shape)
print("X_test.shape ", X_test.shape)
print("y_train.shape ", y_train.shape)
print("y_test.shape ", y_test.shape)
```

### ML(SVM, DT, RF, LR)

### Static : 0, Dynamic : 1

```python
# Classification

from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# SVM

clf_svm = SVC(random_state=0)
clf_svm.fit(X_train, y_train)
pred_svm = clf_svm.predict(X_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))

# confusion matrix

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("SVC Confusion Matrix")
sns.heatmap(confusion_matrix(y_test, pred_svm), cbar=False, linewidths=1.0, annot=True, fmt="d", cmap="Blues")
plt.show()
```

```python
# DT

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, y_train)
pred_dt = clf_dt.predict(X_test)

print("\n--- Decision Tree Classifier ---")
print(accuracy_score(y_test, pred_dt))
print(confusion_matrix(y_test, pred_dt))

# confusion matrix

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("Decision Tree Confusion Matrix")
sns.heatmap(confusion_matrix(y_test, pred_dt), cbar=False, linewidths=1.0, annot=True, fmt="d", cmap="Blues")
plt.show()
```

```python
# RF

rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
pred_rf = rf_clf.predict(X_test)

print("\n--- Random Forest ---")
print(accuracy_score(y_test,pred_rf))
print(confusion_matrix(y_test, pred_rf))

# confusion matrix

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("Random Forest Confusion Matrix")
sns.heatmap(confusion_matrix(y_test, pred_rf), cbar=False, linewidths=1.0, annot=True, fmt="d", cmap="Blues")
plt.show()
```

```python
# LR

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)
pred_lr = clf_lr.predict(X_test)

print("\n--- Logistic Regression Classifier ---")
print(accuracy_score(y_test, pred_lr))
print(confusion_matrix(y_test, pred_lr))

# confusion matrix

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("Logistic Regression Confusion Matrix")
sns.heatmap(confusion_matrix(y_test, pred_lr), cbar=False, linewidths=1.0, annot=True, fmt="d", cmap="Blues")
plt.show()
```
